Subir primero base de datos

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('DataBase.csv')
```

Eliminar los a침os 1960-2001 para reducir el espacio muestral y centrarnos en el presente siglo.

```python

columnas_a_eliminar = [f'{year} [YR{year}]' for year in range(1960, 2001)]

df = df.drop(columns=columnas_a_eliminar)

print (df)

```

Realizamos el primer gr치fico con respecto a la poblaci칩n de Afghanistan que accede a electricidad en funci칩n a nivel porcentual

```python
import pandas as pd
import matplotlib.pyplot as plt

data_to_plot = df.iloc[0, 4:10]
plt.figure(figsize=(10, 6))

plt.bar(data_to_plot.index, data_to_plot.values, color='blue')
plt.xlabel('Columnas 5 a 10')
plt.ylabel('Valores')
plt.title('Access to electricity (% of population)of Afghanistan')
plt.show()






```

Realizamos un segundo gr치fico para el caso de acceso a la electricidad rural en Afganistan a nivel porcentual

```python
import pandas as pd
import matplotlib.pyplot as plt

data_to_plot = df.iloc[2, 4:10]

plt.figure(figsize=(10, 6))

plt.bar(data_to_plot.index, data_to_plot.values, color='green')
plt.xlabel('Columnas 5 a 10')
plt.ylabel('Valores')
plt.title('CO2 emissions (metric tons per capita) in Afghanistan')
plt.show()


```
